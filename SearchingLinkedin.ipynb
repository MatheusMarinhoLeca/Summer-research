```python
import requests
import openpyxl
from bs4 import BeautifulSoup

# OAuth token (you need to generate this via LinkedIn OAuth process)
access_token = 'YOUR_ACCESS_TOKEN'

# List of search terms
search_terms = [
    "Algorithmic",
    "Fairness",
    "Bias",
    "Ethics",
    "Responsible",
    "Ethics",
    "Fair"
]

# Function to search LinkedIn for a given keyword
def search_linkedin(keyword):
    headers = {
        'Authorization': f'Bearer {access_token}',
        'X-Restli-Protocol-Version': '2.0.0'
    }
    url = f'https://api.linkedin.com/v2/search?q=keywords&keywords={keyword}'
    
    response = requests.get(url, headers=headers)
    if response.status_code == 200:
        return response.json()
    else:
        print(f"Error: Received status code {response.status_code} for keyword '{keyword}'")
        return {}

# Function to filter results from 2017 onwards
def filter_by_date(items):
    filtered_items = []
    for item in items:
        creation_date = item.get('created', {}).get('time', None)
        if creation_date and creation_date >= 1483228800:  # Timestamp for 2017-01-01
            filtered_items.append(item)
    return filtered_items

# Function to convert HTML to readable text
def html_to_text(html_content):
    soup = BeautifulSoup(html_content, "html.parser")
    return soup.get_text()

# Main function to search for all keywords on LinkedIn
def main():
    results = []
    for term in search_terms:
        print(f"Searching for '{term}' on LinkedIn...")
        result = search_linkedin(term)
        if result:
            filtered_items = filter_by_date(result.get('elements', []))
            for item in filtered_items:
                post_id = item['id']
                post_url = f"https://www.linkedin.com/feed/update/{post_id}"
                full_post = html_to_text(item.get('description', {}).get('text', ''))
                creation_date = item.get('created', {}).get('time', None)
                creation_date = datetime.datetime.fromtimestamp(creation_date / 1000).strftime('%Y-%m-%d %H:%M:%S') if creation_date else "N/A"
                
                result_entry = {
                    "Post ID": post_id,
                    "PostURL": post_url,
                    "Full Post": full_post,
                    "Date": creation_date,
                    "Topic": "LinkedIn"
                }
                results.append(result_entry)
                
                # Print the results in the console
                print(f"Post ID: {result_entry['Post ID']}")
                print(f"Post URL: {result_entry['PostURL']}")
                print(f"Full Post: {result_entry['Full Post']}")
                print(f"Date: {result_entry['Date']}")
                print(f"Topic: {result_entry['Topic']}")
                print("\n" + "-"*80 + "\n")
        else:
            print(f"No results found for '{term}' on LinkedIn\n")

    # Save results to an Excel file
    wb = openpyxl.Workbook()
    ws = wb.active
    ws.title = "LinkedIn Search Results"
    headers = ["Post ID", "PostURL", "Full Post", "Date", "Topic"]
    ws.append(headers)
    
    for result in results:
        ws.append([
            result["Post ID"],
            result["PostURL"],
            result["Full Post"],
            result["Date"],
            result["Topic"]
        ])

    wb.save("linkedin_search_results.xlsx")
    print("Results have been saved to 'linkedin_search_results.xlsx'.")

if __name__ == "__main__":
    main()
```
